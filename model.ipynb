```python
train_copy = df_train.copy()
train_copy = train_copy.replace(-1, np.NaN)
```

```python
for i in train_copy.columns:
    if train_copy[i].isnull().sum() > 5000:
        train_copy.drop(i, inplace=True, axis=1)
```

```python
train_copy.isnull().sum()
```

```python
bin_features = []
for i in train_copy.columns:
    if i.endswith('bin'):
        bin_features.append(i)
```

```python
cat_features = []
for i in train_copy.columns:
    if i.endswith('cat'):
        cat_features.append(i)
```

```python
# Correlation of binary features
plt.figure(figsize=(15,15))
sns.heatmap(train_copy[bin_features].corr(), annot=True)
```

```python
# Look for null values
train_copy[bin_features].isnull().sum()
```

```python
# Count plot

for i in bin_features:
    sns.countplot(train_copy[i], hue=train_copy['target'])
    plt.show()
```

```python
# EDA of categorical features

plt.figure(figsize=(15,15))
sns.heatmap(train_copy[cat_features].corr(), annot=True)
```

```python
for i in cat_features:
    sns.countplot(train_copy[i], hue=train_copy['target'])
    plt.show()
```

```python
train_copy[cat_features].isnull().sum()
```

```python
# Handle missing values:

for i in cat_features:
    if train_copy[i].isnull().sum() >= 1:
        train_copy[i].fillna(train_copy[i].mean(), inplace=True)
```

```python
train_copy[cat_features].isnull().sum()
```

```python
# EDA on continuous features

cont_features = [i for i in train_copy if i not in cat_features+bin_features]
cont_features.pop(1)
cont_features.pop(0)
cont_features
```

```python
train_copy[cont_features].isnull().sum()
```

```python
for i in cont_features:
    sns.distplot(train_copy[i])
    plt.show()
```

```python
for i in train_copy[cont_features]:
    sns.boxplot(x=train_copy['target'], y=train_copy[i])
    plt.show()
```

```python
for i in train_copy[cont_features]:
    if train_copy[i].isnull().sum() >= 1:
        train_copy[i].fillna(train_copy[i].mean(), inplace=True)
```

```python
train_copy[cont_features].isnull().sum()
```

```python
train_copy = train_copy.drop('id', axis=1)
```

```python
train_copy.head()
```

```python
# Split into train and test dataset
X = train_copy.iloc[:,1:]
y = train_copy['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train, X_test)
X_train = scaler.transform(X_train)
X_test = scaler.transform
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=150, max_depth=8, min_samples_leaf=4, max_features=0.2, n_jobs=-1, random_state=0)

rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
importance = rf.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```
